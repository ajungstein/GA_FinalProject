```python
import pandas as pd
```

### First: Load 2015 data


```python
data = pd.read_csv('../data/NFLPlaybyPlay2015.csv', parse_dates=['Date'], low_memory=False)
```

```python
data.info()
```

```python
data_TomBrady = data.query("Passer=='T.Brady' and Touchdown==1")
cols_to_show = ['Date', 'GameID', 'Drive', 'down', 'Passer', 'FirstDown', 'qtr', 'time', 'desc', 'Yards.Gained', 'PlayType']
data_TomBrady_cols = data_TomBrady[cols_to_show]
data_TomBrady_cols.info()
```

```python
pd.set_option('display.max_colwidth', -1)
data_TomBrady_cols.sort_values(['Date', 'qtr', 'time'], ascending=[True, True, False])
```
